**数据产生过程**
- 城市地区的网络被划分成由位置区域代码(LAC)表示的几个大区域。每个LAC包含多个无线网络控制器(RNC)，该控制器由多个与位置相关的基站(BS)或蜂窝塔组成。每个BS与由小区ID(CI)表示的几个扇区天线(通常三个方向为3)相关联。每个小区使用与相邻小区不同的一组频率，以避免干扰并在每个小区内提供有保证的带宽。

- 在MD连接到网络之前，它会测量附近候选扇区的无线电信号强度参数，并连接到**信号强度最强**的最佳扇区。通常，MR数据包含测量参数，例如6个附近扇区的信号强度的无线电信号强度指标(RSSI)。图 1 显示了一个 MR 记录示例，包括用户 ID IMSI（国际移动系统身份）、时间戳、六个 BS ID（CDMA 网络的 RNC ID 或 LAC 和 CI；LTE 网络的 enodeB ID）和六个RSSI = RSCP -EcNo（RSCP 为接收信号代码功率和 EcNo 是每比特能量对噪声功率密度）。表 2 显示了 MR 记录的详细示例。 RNCID_x 和 CellID_x 表示 RNC ID 和 Cell ID 以标识位置相关 BS 中的唯一扇区。 RTT_x 是信号从 MD 到扇区的往返时间。 UE_Rx_Tx_x 是MD接收和发送UE（用户设备）信号的时间。 MRrecord 通常包含来自 6 个附近扇区的值。但在现实世界的电信网络中，超过 50% 的 MR 记录的缺失值不超过两个扇区。除了 RSSI，MR 数据还可以有其他无线电参数测量，例如 TOA/TDOA 和 AOA。不失一般性，我们仅使用**RSSI以及位置相关BS的工程参数来构建基于MR的定位系统**。

**手机信号数据属性含义：**
- MRTime：记录该MR的记录时间
-  RNCID_1和CellID_1构成了唯一标识该服务主基站的编号
- id_combine= RNCID_1+ CellID_1用来辅助判断是否属于同一主基站
-  Longitude和Latitude为手机所在位置的经纬度信息
- LinkNum: 邻接基站个数

每个MR记录包括最多7个非空连接的基站信息，分别对应：
- RNCID_[1-7]和CellID_[1-7]标识唯一基站
- RSSI_[1-7]: RSSI是射频信号理论术语，主要应用于发射机和接收机之间的距离测量，可以代表信号的强度。 该方法是依据接收信号能量强度确定距离，对通信信道参数要求较高。 其测距理论是：依据无线电波或声波在介质中传输，信号功率是随传播距离衰减的原理。
- SignalLevel_[1-7]: 手机信号级别是Wi-Fi信号的强度。提供信号可视化是为了让您更好地了解您的信噪比 (SNR)。

## 2. 工具函数

````python
# 感谢学长的分享！
lonStep_1m = 0.0000105
latStep_1m = 0.0000090201

# 划分栅格
class RoadGrid:
    def __init__(self, label, grid_size):
        length = grid_size * latStep_1m
        width = grid_size * lonStep_1m
        self.length = length
        self.width = width
        def orginal_plot(label):
            tr = np.max(label,axis=0)
            tr[0]+=25*lonStep_1m
            tr[1]+=25*latStep_1m
            # plot(label[:,0], label[:,1], 'b,')
            bl = np.min(label,axis=0)
            bl[0]-=25*lonStep_1m
            bl[1]-=25*latStep_1m

            # width = (tr[1]-bl[1])/100
            # wnum =int(np.ceil((tr[1]-bl[1])/length))
            # for j in range(wnum):
                # hlines(y = bl[1]+length*j, xmin = bl[0], xmax = tr[0], color = 'red')

            # lnum = int(np.ceil((tr[0]-bl[0])/width))
            # for j in range(lnum):
                # vlines(x = bl[0]+width*j, ymin = bl[1], ymax = tr[1], color = 'red')
            return bl[0], tr[0], bl[1], tr[1]

        xl,xr,yb,yt = orginal_plot(label)
        self.xl = xl
        self.xr = xr
        self.yb = yb
        self.yt = yt
        gridSet = set()
        grid_dict = {}
        self.grid_dict = {}
        for pos in label:
            lon = pos[0]
            lat = pos[1]

            m = int((lon-xl)/width)
            n = int((lat-yb)/length)
            if (m,n) not in grid_dict:
                grid_dict[(m,n)] = []
            grid_dict[(m,n)].append((lon, lat))
            gridSet.add((m,n))
        # print len(gridSet)
        gridlist = list(gridSet)

            
            
        grid_center = [tuple(np.mean(np.array(grid_dict[grid]),axis=0)) for grid in gridlist]


        # for gs in gridSet:
            # xlon = xl+gs[0]*width
            # ylat = yb+gs[1]*length
            # bar(xlon,length,width,ylat,color='#7ED321')
        self.gridlist = gridlist

        self.grids = [(xl+i[0]*width,yb + i[1]*length) for i in grid_dict.keys()] # 左下角的点
        self.grid_center = grid_center
        self.n_grid = len(self.grid_center)
        self.grid_dict = grid_dict

    def transform(self, label, sparse=True):
        def one_hot(idx, n):
            a = [0] * n
            a[idx] = 1
            return a
        grid_pos = [self.gridlist.index((int((i[0]-self.xl)/self.width),int((i[1]-self.yb)/self.length))) for i in label]
        if sparse:
            grid_pos = np.array([one_hot(x, len(self.gridlist)) for x in grid_pos], dtype=np.int32)
        return grid_pos
    
def rad(d):
    return d * math.pi / 180.0

# 地理坐标系：为球面坐标。 参考平面地是椭球面，坐标单位：经纬度；
# 投影坐标系：为平面坐标。参考平面地是水平面，坐标单位：米、千米等；
# 地理坐标转换到投影坐标的过程可理解为投影。（投影：将不规则的地球曲面转换为平面）

# 目前国内主要有三种地理坐标系
# 1、WGS84坐标系：即地球坐标系（World Geodetic System），国际上通用的坐标系。
# 设备包含的GPS芯片或者北斗芯片获取的经纬度一般都是为WGS84地理坐标系，目前谷歌地图采用的是WGS84坐标系（中国范围除外）。
# 2、GCJ02坐标系：即火星坐标系，国测局坐标系。是由中国国家测绘局制定。由WGS84坐标系经加密后的坐标系。谷歌中国和搜搜中国采用。
# 3、BD09坐标系：百度坐标系，GCJ02坐标系经加密后的坐标系。

# 投影：墨卡托投影、高斯-克吕格 (Gauss-Krüger) 投影
# 感兴趣的同学可以在https://desktop.arcgis.com/zh-cn/arcmap/10.3/guide-books/map-projections/list-of-supported-map-projections.htm深入了解

# gps两点间距离（单位为米）
def distance(true_pt, pred_pt):
    lat1 = float(true_pt[1])
    lng1 = float(true_pt[0])
    lat2 = float(pred_pt[1])
    lng2 = float(pred_pt[0])
    radLat1 = rad(lat1)
    radLat2 = rad(lat2)
    a = radLat1 - radLat2
    b = rad(lng1) - rad(lng2)
    s = 2 * math.asin(math.sqrt(math.pow(math.sin(a/2),2) +
    math.cos(radLat1)*math.cos(radLat2)*math.pow(math.sin(b/2),2)))
    s = s * 6378.137
    s = round(s * 10000) / 10
    return s
````

**数据缺失类型**
- 随机丢失（MAR，Missing at Random）
随机丢失意味着数据丢失的概率与丢失的数据本身无关，而仅与部分已观测到的数据有关。也就是说，数据的缺失不是完全随机的，该类数据的缺失依赖于其他完全变量。
- 完全随机丢失（MCAR，Missing Completely at Random）
数据的缺失是完全随机的，不依赖于任何不完全变量或完全变量，不影响样本的无偏性。简单来说，就是数据丢失的概率与其假设值以及其他变量值都完全无关。
- 非随机丢失（MNAR，Missing not at Random）
数据的缺失与不完全变量自身的取值有关。分为两种情况：缺失值取决于其假设值（例如，高收入人群通常不希望在调查中透露他们的收入）；或者，缺失值取决于其他变量值（假设女性通常不想透露她们的年龄，则这里年龄变量缺失值受性别变量的影响）。

在**前两种情况**下可以根据其出现情况删除缺失值的数据，同时，随机缺失可以通过已知变量对缺失值进行估计。

在**第三种情况**下，删除包含缺失值的数据可能会导致模型出现偏差，同时，对数据进行填充也需要格外谨慎。

分析可知，MR数据的RSSI数据缺失可能是因为确实没有6个辅助基站，也可能是因为数据获取过程中丢失。【这里一个疑问是邻近基站的个数应该是准确的信息，那么所给数据集中的缺失值是因为确实没有6个辅助基站的数据】，因此，MR数据缺失类型属于MNAR。

**缺失值处理方法**


````python
# 验证临近基站的个数信息与缺失值信息，从而确定数据缺失是由于缺失不存在辅助基站
miss_res = hand_data.isnull().sum(axis=1)

# # 对于每一行，通过列名name访问对应的元素
error_num = 0
for idx,row in hand_data.iterrows():
    if row['linknum'] != 7 - miss_res[idx] / 2:
        # print(idx)
        error_num = error_num + 1
print("缺失值与基站数不对等数量：", error_num)
````

通过查看原始数据集，发现这里的缺失值是连续的，index从1887到2108，共222行数据中，基站的数量都是只有4个，但统计的临近基站却都是5个。因为不是自己采集的数据，猜测一下几个原因：
- 该区域内存在一个基站出现故障，导致无法获取它的数据，但主基站可能会记录它原始临近基站的个数？
- 临近基站的个数是认为根据获取的RSSI和SignalLevel数据添加的，这样的前面的分析就没有意义了，因为数据的缺失值可能是因为认为的丢失，则需要进行插补处理

个人偏向第一种，基站出现故障，因为认为添加应该是程序处理，这种连续的出错应该是不常见的！

对于这些数据，可以直接更改其Linknum，人工修复数据的问题。

````python
# 修复数据中的错误：逻辑如下，手动更改文件了
for idx,row in hand_data.iterrows():
    if row['linknum'] != 7 - miss_res[idx] / 2:
        row['linknum'] = 7 - miss_res[idx] / 2
````

````python
# 验证
error_num = 0
for idx,row in hand_data.iterrows():
    if row['linknum'] != 7 - miss_res[idx] / 2:
        # print(idx)
        error_num = error_num + 1
print("缺失值与基站数不对等数量：", error_num)
````

通过上述分析可以发现，虽然数据中存在缺失值，但这些缺失值并不是人为等主观因素导致的，而是基站本身的客观因素决定的。因此，这些缺失值并不需要进行填充或者删除等处理，而是需要在模型设计中利用好数据的这个特性！

比如，在分类模型中，我们可以将缺失值作为新的分类。偷学一波：“岛屿 - 湖泊 - 椰子树”模型 --随机森林。【https://zhuanlan.zhihu.com/p/164990454】
> 因为行列都随机，才能够做到真正的把整个数据表随机切分成多份，每个模型使用一份，只要模型的数量足够，总有模型能够在最大程度上获取数据集的价值。而且因变量的分类往往又是极不平衡的。至于如何将这些种好的树的信息又再收集，便可以将陆地上比较近的几棵树上面再弄一个收集器，把这几棵树从陆地上收集到的养分再递进一层汇总，最终实现陆地养分汇总于树木，树木养分汇总于收集器，收集器养分汇总于更上层的另一个收集器，最终实现整片数据海洋中多个岛屿的信息汇总，这便是周志华团队和蚂蚁金服的合作的用分布式深度随机森林算法检测套现欺诈。

根据这个思路，我们可以在分类过程中根据临近基站的个数切分为多份，这样不同模型使用一份，只要模型数量足够，总有模型能够在最大程度上获取数据集的价值。不过，目前数据集只有5000多个，这样分类应该没多大的效果。

````python
# 查看grid_id分布情况
import matplotlib.pyplot as plt
from collections import Counter

grid_fenbu = Counter(hand_data['grid_id'])

# print(grid_fenbu)
print("Most (grid_id, num): ", grid_fenbu.most_common(10))
# n least common elements
print("Least (grid_id, num): ", grid_fenbu.most_common()[:-10-1:-1]) 

dic = sorted(grid_fenbu.items(), key=lambda asd: asd[0], reverse=False)

# 将已排序列表包含的元组进行分离
grid_ids = []  # grid_id
num = []  # 出现次数
for key, value in dic:
    if key != 253:   # 因为这类数据有453个，远远多于其它类别中的数据
        grid_ids.append(key)
        num.append(value)


# 统计图
plt.bar(range(len(grid_ids)), num)
plt.title(u'grid fenbu')
plt.show()
````

````python
import math

print("分类算法", "中位数误差", "平均误差", "0.9误差")
for name, clf in classifiers.items():
    # 每种分类算法上进行分组训练
    print(name, ": ")
    med, mea, nin = [], [], []
    for linknum, group in hand_data.groupby('linknum'):
        # linknum： 2 3 4 5 6 7
        # 划分数据集
        X_train,X_test,y_train,y_test = train_test_split(
                group,
                group[['longitude','latitude','grid_id']],
                test_size=0.2,
                random_state=200)    
        regr = clf.fit(X_train[feature_cloumns_name[linknum-2]], y_train['grid_id'])
        pred = regr.predict(X_test[feature_cloumns_name[linknum-2]])
        pred_loc = np.array([grider.grid_center[idx] for idx in pred])
        err = [distance(p,t) for p, t in zip(pred_loc, y_test[['longitude','latitude']].values)]
        err = sorted(err)
        med.append(np.median(err))
        mea.append(np.mean(err))
        nin.append(err[int(len(err)*0.9)])
        print("临近基站数", linknum, ": ", np.median(err), np.mean(err), err[int(len(err)*0.9)])
    # 分类器   中位数  平均误差
    print ("分组累积误差：", np.mean(med), np.mean(mea), np.mean(nin))
````

总结上述训练结果(中位数误差、平均误差、0.9误差)：

| 分类算法 | 一般数据训练(2个基站) | 分组训练 |

进行分组训练后，中位数误差：
- 随机森林： 327.48 ---> 17.79
- K最近邻：402.30 ---> 90.94
- 决策树：392.21 ---> 152.83
- 贝叶斯：445.4 ---> 152.83

这里主要采用中位数误差而不是平均误差评判的主要原因是栅格个划分导致栅格分布中存在较大的数据【栅格划分之后优化】，大量栅格中只有1个数据导致会存在个别分类误差较大从而使平均误差指标意义不大。比如在K近邻分组训练中，中位数误差为90.94，说明一般的分类误差比这个小，但0.9误差却高达826.91，少数的较大误差导致平均误差偏向少量分类结果中的误差。

## 4. 相对位置定位模型

按照每个MR记录对应的主基站对MR记录进行分组，使得每组MR记录都有相同的主基站，总的分组个数即为主基站个数。假定某主基站的经纬度坐标为<x0, y0>，该分组中的某MR记录对应的经纬度坐标为<x, y>，则计算该MR记录与主基站的相对位置为：<x‘ y‘>=< x-x0, y-y0>。在完成每个分组的MR记录相对位置计算之后，针对每个分组构建一个对应的MR定位模型，不过该模型是以MR记录与主基站的相对位置作为标签。使用处理好的训练集用于训练模型，测试数据集用于测试统计，通过上述随机森林模型预测测试数据的相对位置<x‘’，y‘‘>，然后计算还原为原始位置：<x‘’+ x0，y‘‘+ y0>；比较和讨论a)和c）方法之间的优劣。

````python
import pandas as pd
import numpy as np
````

````python
# 读取数据
hand_data = pd.read_csv("./handledata/2G_data_hand.csv", encoding = 'gbk')

#匹配工参 -- 用来计算相对数据
hand_data_gongcan = pd.read_csv("./data/2G_gongcan.csv", encoding = 'gbk')

# 将列名全部小写
hand_data.rename(str.lower, axis='columns', inplace=True)
hand_data_gongcan.rename(str.lower, axis='columns', inplace=True)

# 重命名
hand_data.rename(columns={'rncid_1':'rncid','cellid_1':'cellid'}, inplace=True)
hand_data_gongcan.rename(columns={'longitude':'real_longi','latitude':'real_lati'}, inplace=True)

print("数据集：")
print(hand_data.head(1))
print("工参：")
print(hand_data_gongcan.head(1))
````

````python
# 合并两个DF
data = pd.merge(hand_data,hand_data_gongcan,how="left",on=["rncid","cellid"])

print(data.head(1))
````

````python
# 按照基站进行分组
print(data.groupby(['rncid','cellid'])['cellid'].count())
````

根据主基站共分为43个组，其中每个组的数组分布如上所示。

````python
# 计算相对位置
relative_longitude = []   
relative_latitude = []
for index, row in data.iterrows():
    # 手机坐标减去基站坐标
    relative_longitude = row['longitude'] - row['real_longi']
    relative_latitude = row['latitude'] - row['real_lati']

data['relative_longi'] = relative_longitude
data['relative_lati'] = relative_latitude

print(data.head(1))
````

````python
# 回归预测函数

from sklearn.tree import DecisionTreeRegressor       # 决策树
from sklearn.ensemble import RandomForestRegressor   # 随机森林
from sklearn.model_selection import train_test_split  # 数据集划分

# 分类器
regressors = {
    "Random Forest": RandomForestRegressor(n_estimators=100),
    "K Neighbors" : KNeighborsRegressor(n_neighbors=2),
    "Decision Tree": DecisionTreeRegressor()
}
````

````python
# 缺失值处理
data.dropna(subset=['real_longi'],inplace=True)
````

````python
# linknum=2
feature_2_cloumns_name = [ 'signallevel_1','rssi_1', 
                        'signallevel_2','rssi_2'
                        ]

# linknum=7
feature_7_cloumns_name = [ 'signallevel_1','rssi_1', 
                        'signallevel_2','rssi_2', 
                        'signallevel_3','rssi_3',
                        'signallevel_4','rssi_4', 
                        'signallevel_5','rssi_5',
                        'signallevel_6','rssi_6', 
                        'signallevel_7','rssi_7',
                        'linknum'
                        ]
````

````python
# 一般性 数据训练
# 按照主基站进行分组，每一组使用随机森林进行回归预测
import numpy as np
import math

for name, model in regressors.items():
    print(name + ": ")
    # 分组进行回归算法
    med, mea, nin = [], [], []
    for idx, group in data.groupby(['rncid','cellid']):
        group.fillna(-99)
        # 划分数据集
        X_train,X_test,y_train,y_test = train_test_split(
                group,
                group[['real_longi','real_lati','relative_longi','relative_lati']],
                test_size=0.2,
                random_state=200)
        model.fit(X_train[feature_7_cloumns_name], y_train[['relative_longi','relative_lati']].values)
        pred = model.predict(X_test[feature_7_cloumns_name])
        # 换回预测手机定位坐标  预测相对坐标+基站坐标
        pre_loc = []   # 预测坐标数据
        real_loc = []  # 真实坐标数据
        for index, row in X_test.iterrows():
            pre_loc.append([row['real_longi'],row['real_lati']])
            real_loc.append([row['longitude'],row['latitude']])
        for i in range(len(pre_loc)):
            pre_loc[i][0] = pre_loc[i][0] + pred[i][0]
            pre_loc[i][1] = pre_loc[i][1] + pred[i][1]
#         print("预测坐标：", pre_loc[0])
#         print("真实坐标：", real_loc[0])
        # 计算误差
        err = [distance(p,t) for p, t in zip(pre_loc, real_loc)]
        err = sorted(err)
        med.append(np.median(err))
        mea.append(np.mean(err))
        nin.append(err[int(len(err)*0.9)])
    # 分类器   中位数  平均误差
    print(np.mean(med), np.mean(mea), np.mean(nin))
        
````
